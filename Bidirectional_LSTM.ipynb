**Bidirectional LSTMs**

Bidirectional LSTMs are a type of recurrent neural network (RNN) architecture specifically designed for processing sequential data like text, time series, or audio signals. They address a limitation of standard LSTMs, which only process information in the forward direction (left to right). This can be insufficient for tasks where understanding the context from both sides of a sequence is crucial.

**How Bidirectional LSTMs Work**

Bidirectional LSTMs overcome this limitation by incorporating two LSTMs:

- <font color='green'>**Forward LSTM:**</font> Processes the sequence in the normal direction, capturing dependencies between earlier and later elements.
- <font color='green'>**Backward LSTM:**</font> Processes the sequence in the reverse direction, capturing dependencies between later and earlier elements.

This bidirectional processing allows Bidirectional LSTMs to leverage context from both sides of a sequence, leading to a more comprehensive understanding of the data.

**Advantages of Bidirectional LSTMs over LSTMs**

- <font color='green'>**Improved Contextual Understanding:**</font> Bidirectional LSTMs can better grasp the relationships between elements in a sequence by considering information from both directions. This is particularly beneficial for tasks like sentiment analysis (understanding the overall sentiment of a sentence) or machine translation (accurately capturing the meaning and context across languages).
- <font color='green'>**Modeling Long-Term Dependencies:**</font> Bidirectional LSTMs are adept at capturing long-range dependencies within sequences, which can be crucial for tasks where distant elements influence each other. For example, in speech recognition, the meaning of a word might depend on words spoken several seconds earlier.

**While it's true that all models, including AI models, have limitations and can be fooled by adversarial examples, Bidirectional LSTMs offer significant advantages over LSTMs in many sequential data processing tasks.**

```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
###Drop Nan Values
df=df.dropna()
```

```python
## Get the Independent Features

X=df.drop('label',axis=1)
```

```python
## Get the Dependent features
y=df['label']
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Bidirectional
from tensorflow.keras.layers import Dropout
```

```python
### Vocabulary size
voc_size=5000
```

```python
messages=X.copy()
```

```python
messages['title'][1]
```

```python
messages.reset_index(inplace=True)
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
### Dataset Preprocessing
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus = []
for i in range(0, len(messages)):
    print(i)
    review = re.sub('[^a-zA-Z]', ' ', messages['title'][i])
    review = review.lower()
    review = review.split()

    review = [ps.stem(word) for word in review if not word in stopwords.words('english')]
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr=[one_hot(words,voc_size)for words in corpus]
onehot_repr
```

```python
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
embedded_docs[0]
```

```python
## Creating model
embedding_vector_features=40
model=Sequential()
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))
model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
## Creating model
embedding_vector_features=40
model1=Sequential()
model1.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model1.add(Bidirectional(LSTM(100)))
model1.add(Dropout(0.3))
model1.add(Dense(1,activation='sigmoid'))
model1.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model1.summary())
```

```python
len(embedded_docs),y.shape
```

```python
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
```

```python
X_final.shape,y_final.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
### Finally Training
model1.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
y_pred1=model1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python

y_test = y_test.astype(np.int64)
y_pred1 = y_pred1.astype(np.int64)

# Recalculate the confusion matrix
confusion_matrix = confusion_matrix(y_test, y_pred1)

# Print the confusion matrix
print(confusion_matrix)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred1)
```

```python

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred1))
```
